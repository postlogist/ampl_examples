```python
import pandas as pd
from geocoder import geocode_df
```

```python
print(geocode_df.__doc__)
```

Чтобы посмотреть справку по использованию функции, введите в кодовой ячейке: `geocode_df?`

Загрузим список городов (если названия дублируются в разных регионах, то можно уточнить запрос, добавив столбец регион и страну).

```python
locations = pd.read_excel("locations.xlsx")

print(f"Число строк: {len(locations)}")
locations.head()
```

Геокодирование (только по городу):

```python
result = geocode_df(
    locations,
    address=None,
    city="city",
    region=None,
    country=None,
    user_agent="emm_geocode_example",   # смените на название своего проекта
    min_delay_seconds=1.1,
    max_retries=3,
    language="ru",
)
```

```python
result
```

```python
output_file = "locations_geocoded.xlsx"

result.to_excel(output_file, index=False)
print(f"\nСохранено в {output_file}")
```
